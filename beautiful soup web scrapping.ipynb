```python
import requests
from bs4 import BeautifulSoup
```

```python
Country = 'Hong+Kong'
job = 'data+science'
u = "https://wuzzuf.net/search/jobs?a=spbg&q=machine%20learning"
page = requests.get(u)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
j_t = soup.find_all("h2",class_="css-m604qf")
for i in j_t:
    print(i.text)
```

```python
loc = soup.find_all("span",class_="css-5wys0k")
for i in loc:
    print(i.text)
```

```python
company = soup.find_all("a",class_="css-17s97q8")
for i in company:
    print(i.text)
```

```python
desc= soup.find_all("div",class_="css-y4udm8")
for d in desc:
    print(d.text)
```
